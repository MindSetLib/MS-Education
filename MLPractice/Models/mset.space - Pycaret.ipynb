```python
import pandas as pd
from pycaret.regression import *
```

```python
train = pd.read_csv('./data/train.csv',index_col=0)
test = pd.read_csv('./data/test.csv',index_col=0)
```

```python
exp1 = setup(train, target = 'loss', silent=True)
```

```python
lightgbm = create_model('lightgbm')
```

```python
catboost = create_model('catboost')
```

```python
xgboost = create_model('xgboost')
```

```python
tuned_lightgbm = tune_model(lightgbm)
```

```python
tuned_catboost = tune_model(catboost)
```

```python
tuned_xgboost = tune_model(xgboost)
```

```python
blender_specific = blend_models(estimator_list = [lightgbm,catboost,xgboost,tuned_lightgbm ,tuned_catboost,tuned_xgboost])
```

```python
predictions = predict_model(blender_specific, data = test)
```

```python
predictions
```

```python
preds_final = pd.DataFrame()
preds_final['loss'] = predictions['Label']
preds_final['id']  = test.index.copy()
```

```python
preds_final.to_csv('pycaret.csv', index=False)
```
