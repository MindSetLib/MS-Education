```python
loan_status_dict={
    'Fully Paid' : 0,
    'Charged Off': 1,
    'Current':1,
    'Late (31-120 days)':1,
    'In Grace Period':1,
    'Late (16-30 days)':1,
    'Default':1
}
```

```python
df['Target'] = df['loan_status'].map(loan_status_dict)
```

```python
X = df[['loan_amnt',
 'funded_amnt',
 'funded_amnt_inv',
 'term',
 'int_rate',
 'installment',
 'grade']].copy()
```

```python
y = df['Target'].copy()
```

```python
categorical_columns=[c for c in X.columns if X[c].dtype.name == 'object']
numerical_columns=[c for c in X.columns if X[c].dtype.name != 'object']
```

```python
data_category_encoded = pd.get_dummies(X[categorical_columns]) 
```

```python
data_numerical = (X[numerical_columns]-X[numerical_columns].mean())/X[numerical_columns].std()
```

```python
X = pd.concat([data_category_encoded, data_numerical], axis=1)
```

```python
X
```

```python
X
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.metrics import roc_auc_score, accuracy_score, confusion_matrix, f1_score, precision_score, recall_score, roc_curve
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
skf = StratifiedKFold(n_splits=5, shuffle=False)
cv_scores = cross_val_score(lr, X, y, cv=skf, n_jobs=10, scoring = 'roc_auc')
print(cv_scores)
print(cv_scores.mean())
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=False)
```

```python
cv_scores = cross_val_score(lr, X, y, cv=skf, n_jobs=10, scoring = 'roc_auc')
```

```python
print(cv_scores)
cv_scores.mean()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_prob=lr.predict_proba(X_test)[:,1]
y_pred=lr.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
fpr, tpr, th = roc_curve(y_test, y_pred_prob)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (6, 6)
plt.plot(fpr, tpr, 'r');
plt.plot([0, 1], [0, 1],'k--');
plt.xlim([0, 1]);
plt.ylim([0, 1]);
plt.ylabel('True Positive Rate');
plt.xlabel('False Positive Rate');
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
skf = StratifiedKFold(n_splits=5, shuffle=False)
cv_scores = cross_val_score(rf, X, y, cv=skf, n_jobs=10, scoring = 'roc_auc')
print(cv_scores)
print(cv_scores.mean())
y_pred_prob=rf.predict_proba(X_test)[:,1]
y_pred=rf.predict(X_test)
fpr, tpr, th = roc_curve(y_test, y_pred_prob)
plt.rcParams['figure.figsize'] = (6, 6)
plt.plot(fpr, tpr, 'r');
plt.plot([0, 1], [0, 1],'k--');
plt.xlim([0, 1]);
plt.ylim([0, 1]);
plt.ylabel('True Positive Rate');
plt.xlabel('False Positive Rate');
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
skf = StratifiedKFold(n_splits=5, shuffle=False)
cv_scores = cross_val_score(dt, X, y, cv=skf, n_jobs=10, scoring = 'roc_auc')
print(cv_scores)
print(cv_scores.mean())
y_pred_prob=dt.predict_proba(X_test)[:,1]
y_pred=dt.predict(X_test)
fpr, tpr, th = roc_curve(y_test, y_pred_prob)
plt.rcParams['figure.figsize'] = (6, 6)
plt.plot(fpr, tpr, 'r');
plt.plot([0, 1], [0, 1],'k--');
plt.xlim([0, 1]);
plt.ylim([0, 1]);
plt.ylabel('True Positive Rate');
plt.xlabel('False Positive Rate');
```
